```python
import pandas as pd
import sqlite3

crime_data = pd.read_csv('/Crimes_-_2001_to_Present.csv')
conn = sqlite3.connect('db.db')
crime_data.to_sql("crime_data", conn)

# Close the connection
conn.close()
```

```python
%sql sqlite:///db.db
```

```python
%%sql
SELECT *
FROM crime_data
LIMIT 5;
```

```python
%%sql
SELECT "Primary Type", COUNT(*) AS count
FROM crime_data
GROUP BY "Primary Type"
ORDER BY count DESC
LIMIT 10;
```

```python
%%sql
SELECT COUNT(*) FROM crime_data;
```

```python
%%sql
SELECT DISTINCT "Primary Type" FROM crime_data;
```

```python
%%sql
SELECT "Primary Type", COUNT(*) AS count
FROM crime_data
GROUP BY "Primary Type"
ORDER BY count DESC;
```

```python
%%sql
SELECT Year,
       COUNT(*) AS num_incidents,
       AVG(1.0 * COUNT(*)) OVER () AS avg_incidents_per_year,
       MIN(COUNT(*)) OVER () AS min_incidents_per_year
FROM crime_data
GROUP BY Year
HAVING num_incidents > 1000
ORDER BY Year;
```
